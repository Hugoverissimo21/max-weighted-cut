```python
df.set_index([("Graph", "name"), ("Graph", "n"), ("Graph", "m"),
              ("Exhaustive", "#operations"), ("Exhaustive", "#solutions tested"),
              ("Heuristic", "#operations"), ("Heuristic", "#solutions tested")], inplace=True)
df
```

```python
index_list = df.index.tolist()

def minNan(x):
    x1 = [i for i in x if not pd.isnull(i)]
    if x1 == []:
        return x[0]
    else:
        return min(x1)

for index in index_list:
    locked = df.loc[index]
    if len(locked) == 1:
        pass
    else:
        for col in locked.columns:
            locked[col] = minNan(list(locked[col]))
        df.loc[index] = locked.iloc[0].to_frame().T.values[0]
```
